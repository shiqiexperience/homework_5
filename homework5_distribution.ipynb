```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import (uniform, norm, expon, poisson, binom)
```

# Равномерное распределение

```python
p = np.linspace(-2, 3, 100)

plt.plot(p, uniform.pdf(p, 0, 1))
plt.title('График плотности равномерного распределения')
plt.xlabel('x')
plt.ylabel('rho(x)')
plt.grid()
```

```python
plt.plot(x, uniform.cdf(x))
plt.title('График функции равномерного распределения')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
```

# Биномиальное распределение

```python
n=10
p=0.5
x = np.arange(0, n+1)
plt.plot(x, binom.cdf(x, n, p))
plt.title('График Функции распределения биномиального распределения')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
plt.scatter(x, binom.pmf(x, n, p))
plt.xlabel('x')
plt.ylabel('rho(x)')
plt.title('График плотности вероятности биномиального распределения')
plt.grid()
```

# Распределение Пуассона

```python
from scipy.stats import poisson
mu=5
x = np.linspace(0, 19, 10)

plt.plot(x, poisson.cdf(x, mu))
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('График функции распределения Пуассона')
plt.grid()
```

```python
plt.scatter(x, poisson.pmf(x, mu))
plt.xlabel('x')
plt.ylabel('rho(x)')
plt.title('График плотности вероятности распределения Пуассона')
plt.grid()
```

#  Нормальное распределение

```python
x = np.linspace(-5, 5, 100)

plt.plot(x, norm.cdf(x))
plt.xlabel('x')
plt.ylabel('rho(x)')
plt.title('График функции нормального распределения')
plt.grid()
```

```python
plt.plot(x, norm.pdf(x, 0, 2))
plt.xlabel('x')
plt.ylabel('rho(x)')
plt.title('График плотности вероятности нормального распределения')
plt.grid()
```

# Экспоненциальное распределение

```python
from scipy.stats import expon
x = np.linspace(-5, 10, 100)
mu = 3

plt.plot(x, expon.cdf(x, mu))
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('График функции экспоненциального распределения')
plt.grid()
```

```python
plt.plot(x, expon.pdf(x, mu))
plt.xlabel('x')
plt.ylabel('rho(x)')
plt.title('График плотности вероятности экспоненциального распределения')
plt.grid()
```
